```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression

data = pd.read_excel("数据.xlsx")

# 根据行数生成季度索引（从 2011Q1 开始，往后排 len(data) 个季度）
periods = pd.period_range(start="2011Q1", periods=len(data), freq="Q")
data.index = pd.PeriodIndex(periods=len(data), freq="Q")
data.index.name = "时间"
print(data)
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
data = data.infer_objects(copy=False)
data_filled = data.interpolate(method="linear", limit_direction="both")
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_filled)
pca = PCA()
pca.fit(data_scaled)
print(pca.explained_variance_ratio_)

# 第一主成分作为综合指标
pc1 = pca.transform(data_scaled)[:, 0]

# 归一化到0-100
pc1_norm = (pc1 - pc1.min()) / (pc1.max() - pc1.min()) * 100
```
